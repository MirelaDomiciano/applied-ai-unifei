## Similaridade de Cosseno

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

i  = 0
for grupo, sub_df in atividades_df.groupby(['sala','atividade']):
  i+=1
  if (i < 5):
    continue
  display(Markdown(f'## {grupo[0]} - {grupo[1]}'))
  # vetoriza
  vectorizer = TfidfVectorizer()
  # tokeniza notebooks
  tfidf_matrix = vectorizer.fit_transform(sub_df['corpo'])
  # matrix de similaridades
  similarity_matrix = cosine_similarity(tfidf_matrix)
  # obter a triangular inferior da matrix
  similar = np.tril(similarity_matrix, k=-1)
  # média de similaridade dos trabalhos da triangular inferior
  print(np.mean(similar[similar > 0]))
  print(similar)
  break
```

```python
# Similaridade com TheFuzz
def fuzzy_similarity(s1, s2):
    return fuzz.token_sort_ratio(s1, s2) / 100.0

# Similaridade com LSTM Siamese
def create_siamese_model(input_dim):
    input_a = Input(shape=(input_dim,))
    input_b = Input(shape=(input_dim,))
    shared_lstm = Dense(128, activation='relu')
    encoded_a = shared_lstm(input_a)
    encoded_b = shared_lstm(input_b)
    L1_layer = Lambda(lambda tensors: tf.abs(tensors[0] - tensors[1]))
    L1_distance = L1_layer([encoded_a, encoded_b])
    prediction = Dense(1, activation='sigmoid')(L1_distance)
    model = Model(inputs=[input_a, input_b], outputs=prediction)
    model.compile(loss='binary_crossentropy', optimizer=Adam())
    return model

# Similaridade com Transformers
model_transformer = SentenceTransformer('paraphrase-MiniLM-L6-v2')
def transformer_similarity(s1, s2):
    emb1 = model_transformer.encode(s1, convert_to_tensor=True)
    emb2 = model_transformer.encode(s2, convert_to_tensor=True)
    return float(util.pytorch_cos_sim(emb1, emb2)[0][0])
```

```python
resultados = []

for grupo, sub_df in atividades_df.groupby(['sala', 'atividade']):
    textos = list(sub_df['corpo'])
    n = len(textos)
    for i in range(n):
        for j in range(i + 1, n):
            t1, t2 = textos[i], textos[j]

            # Coseno
            vectorizer = TfidfVectorizer()
            tfidf_matrix = vectorizer.fit_transform([t1, t2])
            cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

            # Fuzz
            fuzz_sim = fuzzy_similarity(t1, t2)

            # Siamese (usar embeddings ou simplificação com TF-IDF)
            siam_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]  # substituto simplificado

            # Transformer
            trans_sim = transformer_similarity(t1, t2)

            resultados.append({
                'sala': grupo[0],
                'atividade': grupo[1],
                'par': f'{i}-{j}',
                'cosseno': cos_sim,
                'fuzz': fuzz_sim,
                'siamese': siam_sim,
                'transformer': trans_sim
            })
```

```python
df_resultados = pd.DataFrame(resultados)
medias = df_resultados.groupby(['sala', 'atividade'])[['cosseno', 'fuzz', 'siamese', 'transformer']].mean().reset_index()
media_geral = medias[['cosseno', 'fuzz', 'siamese', 'transformer']].mean()
tecnica_menor_sim = media_geral.idxmin()
valor_menor_sim = media_geral.min()

print("Técnica com menor similaridade média:", tecnica_menor_sim)
print("Valor da similaridade média:", valor_menor_sim)
```

```python
def top_similares_todas_tecnicas(df_resultados, tecnicas=['cosseno', 'fuzz', 'siamese', 'transformer'], top_n=3):
    top_geral = []

    for tecnica in tecnicas:
        for grupo, sub_df in df_resultados.groupby(['sala', 'atividade']):
            top = sub_df.sort_values(by=tecnica, ascending=False).head(top_n).copy()
            top['sala'] = grupo[0]
            top['atividade'] = grupo[1]
            top['tecnica'] = tecnica
            top['valor_similaridade'] = top[tecnica]
            # Apenas as colunas relevantes
            top_geral.append(top[['sala', 'atividade', 'par', 'tecnica', 'valor_similaridade']])

    return pd.concat(top_geral).sort_values(by=['sala', 'atividade', 'tecnica', 'valor_similaridade'], ascending=[True, True, True, False]).reset_index(drop=True)
```

```python
top_similares_df = top_similares_todas_tecnicas(df_resultados, top_n=3)
display(top_similares_df)
```
